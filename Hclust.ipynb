```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_iris
iri=load_iris()
```

```python
irid=pd.DataFrame(iri.data)
```

```python
irid.columns=iri.feature_names
```

```python
irid
```

```python
from sklearn .model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X=scaler.fit_transform(irid)
```

```python
X
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
```

```python
xpca=pca.fit_transform(X)
```

```python
plt.scatter(xpca[:,0],xpca[:,1],c=iri.target)
```

```python
plt.scatter(xpca[:,0],xpca[:,1])
```

```python
import scipy.cluster.hierarchy as cs
```

```python
cs.dendrogram(cs.linkage(xpca,method='complete'))
```

```python
cs.dendrogram(cs.linkage(xpca,method='ward'))
```

```python
cs.dendrogram(cs.linkage(xpca,method='single'))
```

```python
cs.dendrogram(cs.linkage(xpca,method='average'))
```

```python
from sklearn.cluster import AgglomerativeClustering

# Create the Agglomerative Clustering model with Ward's linkage
clust = AgglomerativeClustering(n_clusters=2, linkage='ward')

# Fit the model to the PCA-transformed data
clust.fit(xpca)
```

```python
clust.labels_
```

```python
plt.scatter(xpca[:,0],xpca[:,1],c=clust.labels_)
```
